7. What is the distribution of cancer stages?

```python
import matplotlib.pyplot as plt
import seaborn as sns
stage_order = ['Stage I', 'Stage II', 'Stage III', 'Stage IV']
stage_counts = df['Cancer_stage'].value_counts()
stage_counts = stage_counts.reindex(stage_order)

# Plotting
plt.figure(figsize=(8, 6))
sns.barplot(x=stage_counts.index, y=stage_counts.values, palette='Set2')

# Title and axis labels
plt.title("Distribution of Cancer Stages (Bar Chart)")
plt.xlabel("Cancer Stage")
plt.ylabel("Number of Patients")

# Add value labels on top of bars
for i, value in enumerate(stage_counts.values):
    plt.text(i, value + 1000, str(value), ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

```


Conclusion:

Cancer stage distribution is balanced, with almost equal numbers of patients in each stage (Stage IIV).

10. How has the average treatment delay (in days) changed over the years?


```python
df['Treatment_delay'] = (df['End_treatment_date'] - df['Diagnosis_date']).dt.days
df['Diagnosis_Year'] = df['Diagnosis_date'].dt.year
avg_delay = df.groupby('Diagnosis_Year')['Treatment_delay'].mean().reset_index()
print(avg_delay)
import seaborn as sns
import matplotlib.pyplot as plt

sns.lineplot(data=avg_delay, x='Diagnosis_Year', y='Treatment_delay', marker='o')
plt.title('Average Treatment Delay per Year')
plt.ylabel('Delay (days)')
plt.xlabel('Year')
plt.grid(True)
plt.show()

```

Conclusion:

The average treatment delay for lung cancer patients remained remarkably steady from 2014 to 2024, with only minor yearly fluctuations of less than 2 days.

11. How does family history affect survival?


```python
#  1 = survived, 0 = not survived
# Calculate survival rate by family history
survival_by_history = df.groupby('Family_history')['Survived'].mean().reset_index()
print(survival_by_history)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.barplot(data=survival_by_history, x='Family_history', y='Survived', palette='Set2')
plt.title('Survival Rate by Family History')
plt.ylabel('Survival Rate')
plt.xlabel('Family History of Cancer')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

Conclusion:

Patients with a family history of cancer had a slightly higher survival rate compared to those without, though the difference is minimal.

12. Which countries have the highest survival rates?


```python
country_survival = df.groupby('Country')['Survived'].mean().reset_index()
print(country_survival)
import seaborn as sns
import matplotlib.pyplot as plt
top_countries = country_survival.sort_values(by='Survived', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(data=top_countries, x='Survived', y='Country', palette='viridis')
plt.title('Top 10 Countries by Survival Rate')
plt.xlabel('Survival Rate')
plt.ylabel('Country')
plt.xlim(0, 1)
plt.tight_layout()
plt.show()
```

Conclusion:

Countries like Estonia, Czech Republic, and Netherlands show slightly higher survival rates, but differences are minimal across all countries.

13. What is the distribution of co-morbidities (Hypertension, Asthma, Cirrhosis, Other_cancer)?


```python
comorbidity_cols = ['Hypertension', 'Asthma', 'Cirrhosis', 'Other_cancer']
comorbidity_counts = df[comorbidity_cols].sum().sort_values(ascending=False)
comorbidity_counts.plot(kind='barh', color='lightcoral')
plt.title('Comorbidities among Patients')
plt.xlabel('Number of Patients')
plt.tight_layout()
plt.show()
```

Conclusion:

Hypertension is the most common co-morbidity among lung cancer patients.

14. What is the relationship between smoking status and cancer stage?


```python
smoke_stage_ct = pd.crosstab(df['Smoking_status'], df['Cancer_stage'], normalize='index')
import matplotlib.pyplot as plt
smoke_stage_ct.plot(kind='bar', stacked=True, colormap='Set3', figsize=(8,6))
print(smoke_stage_ct)
plt.title('Cancer Stage Distribution by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Proportion of Patients')
plt.legend(title='Cancer Stage')
plt.tight_layout()
plt.show()
```

Conclusion:

Cancer stage distribution is nearly uniform across all smoking statuses, suggesting that smoking history has no strong influence on the stage at which lung cancer is diagnosed in this dataset.

15. How does the survival rate change with both smoking status and cancer stage?


```python
survival_smoking_stage = df.groupby(['Smoking_status', 'Cancer_stage'])['Survived'].mean().reset_index()
print(survival_smoking_stage)
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.barplot(data=survival_smoking_stage, x='Cancer_stage', y='Survived', hue='Smoking_status', palette='Set1')
plt.title('Survival Rate by Smoking Status Across Cancer Stages')
plt.xlabel('Cancer Stage')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.tight_layout()
plt.legend(title='Smoking Status')
plt.show()
```

Conclusion:

Never smokers and passive smokers show slightly higher survival rates, especially in Stage II and Stage IV. In contrast, current and former smokers consistently have lower survival across all stages.
